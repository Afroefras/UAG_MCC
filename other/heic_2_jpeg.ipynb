```python
from pathlib import Path, PosixPath
from PIL.Image import open as open_img
from pillow_heif import register_heif_opener

class HeicToJpg:
    def __init__(self, imgs_dir: str) -> None:
        self.imgs_dir = Path(imgs_dir)
        register_heif_opener()

    def set_new_directory(self) -> None:
        dir_name = self.imgs_dir.stem
        self.converted_dir = self.imgs_dir.parent.joinpath(f"converted_{dir_name}")
        self.converted_dir.mkdir(exist_ok=True)

    def heic_to_jpg(self, img_dir: PosixPath, verbose: bool) -> None:
        img_name = img_dir.stem
        img = open_img(str(img_dir))

        converted_img_dir = self.converted_dir.joinpath(f"converted_{img_name}.jpeg")
        img.save(converted_img_dir, format="jpeg")
        
        if verbose:
            print(f"{img_name}.heic converted to jpeg!")

    def convert(self, verbose: bool=False) -> None:
        self.set_new_directory()
        
        # for img in self.imgs_dir.glob('*.heic'):
        for img in self.imgs_dir.glob('*.HEIC'):
            self.heic_to_jpg(img, verbose)
```

```python
IMGS_DIR = "/Users/efrainflores/Desktop/convert"

heic_to_jpg = HeicToJpg(IMGS_DIR)
heic_to_jpg.convert(verbose=True)
```
